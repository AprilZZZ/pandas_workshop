### Data Visualization using Matplotlib

A picture speaks a thousand words. Humans understand visual things better. Visualization helps to present things to any kind of audience and can easily explain a complex phenomenon in layman's terms. 

**Matplotlib** is the most popular Python module for data visualization. It is a base library for most of the advanced Python visualization modules, such as Seaborn.

Different types of plots:

- plt.plot() - a line plot is a chart that displays a line between two variables. It has a sequence of data points joined by a segmentðŸ’•
- plt.scatter() - represent the relationship between two numerial values
- plt.pie() - is a circular graph that is split up into wedge-shaped pieces. Each piece is proportionate to the value it represents.
- plt.bar() - a visual tool to compare the values of various groups
- plt.hist() - shows the distribution of a numeric variable. It shows the probability distribution of a continuous variable. A histogram only works on a single variable while a bar graph works on two variables:

Plot features:

- plt.legend()
- plt.title()
- plt.xlabel()
- plt.ylabel()
- plt.figure()
- plt.xticks()

```python
# a histogram only works on a single variable while a bar graph works on two variables
# Create the data
employee_age = [21, 28, 32, 34, 35, 35, 37, 42, 47, 55]

# Create bins for histogram
bins = [20, 30, 40, 50, 60]

# Plot the histogram
plt.hist(employee_age, bins, rwidth=0.6)

# Add X Label on X-axis
plt.xlabel("Employee Age")

# Add X Label on X-axis
plt.ylabel("Frequency")

# Add title to graph
plt.title("Employee Age Distribution")

# Show the plot
plt.show()
```

```python
# bubble plot shows the third dimension of a plot 
# It shows three numerical values: two values are on the x and y axes 
# and the third one is the size of data points (or bubbles)

# set figure size
plt.figure(figsize=(8, 5))

# create the data
countries = ['Qatar','Luxembourg','Singapore','Brunei','Ireland','Norway','UAE','Kuwait']
populations = [2781682, 604245,5757499,428963,4818690,5337962,9630959,4137312]
gdp_per_capita = [130475, 106705, 100345, 79530, 78785, 74356, 149382, 37000]

# scale GDP per capita income to shoot the bubbles in the graph
scaled_gdp_per_capita = np.divide(gdp_per_capita, 30)

colors = np.random.rand(8)

# draw the scatter diagram
plt.scatter(countries, populations, s=scaled_gdp_per_capita, 
            c=colors, cmap='GnBu',
            edgecolors='blue', alpha=0.5)

# add X Label on X-axis
plt.xlabel("Countries")
# add Y Label on X-axis
plt.ylabel("Population")
# add title to graph
plt.title("Bubble Chart")

# rotate x label for clear visualization
plt.xticks(rotation=45)

plt.show()
```

### Pandas plotting

The pandas library offers the plot() method as a wrapper around the Matplotlib library. The plot() method allows us to create plots directly on pandas DataFrames. The following plot() method parameters are used to create the plots:

- kind: A string parameter for the type of graph, such as line, bar, barh, hist, box, KDE, pie, area, or scatter.
- figsize: This defines the size for a figure in a tuple of (width, height).
- title: This defines the title for the graph.
- grid: Boolean parameter for the axis grid line.
- legend: This defines the legend.
- xticks: This defines the sequence of x-axis ticks.
- yticks: This defines the sequence of y-axis ticks.

```python
df = pd.DataFrame({
            'name':['Ajay','Malala','Abhijeet','Yming','Desilva','Lisa'],
            'age':[22,72,25,19,42,38],
            'gender':['M','F','M','M','M','F'],
            'country':['India','Pakistan','Bangladesh','China','Srilanka','UK'],
            'income':[2500,3800,3300,2100,4500,5500]
            })
# create a scatter plot
df.plot(kind='scatter', x='age', y='income', color='red', title='Age vs Income')
plt.show()
```

```python
# create bar plot
df.plot(kind='bar',x='country', y='income', color='orange')
plt.show()
```

### Advanced visualization using the Seaborn package

Seaborn is an open source Python library for high-level interactive and attractive statistical visualization. Seaborn uses Matplotlib as a base library and offers more simple, easy-to-understand, interactive, and attractive visualizations.

The **lm plot** plots the scatter and fits the regression model on it. A scatter plot is the best way to understand the relationship between two variables. Its output visualization is a joint distribution of two variables.

```python
import seaborn as sns
```

```python
df=pd.DataFrame({'x':[1,3,5,7,9,11],'y':[10,25,35,33,41,59]})
sns.lmplot(x='x', y='y', data=df)
plt.show()
```

```python
# by default, lmplot() fits the regression line. We can remove this by setting the fit_reg parameter as False
sns.lmplot(x='x', y='y', data=df, fit_reg=False)
plt.show()
```

```python
# last_evaluation is the evaluated performance of the employee
# satisfaction_level is the employee's satisfaction level in the company
# left means whether the employee left the company or not
df = pd.read_csv('../Python-Data-Analysis-Third-Edition/Chapter05/HR_comma_sep.csv')
sns.lmplot(x='satisfaction_level', y='last_evaluation', data=df, fit_reg=False, hue='left')
plt.show()
# the third variable 'left' is passed in the hue parameter. The hue property is used for color shade.
```

```python
# barplot() offers the relationship between a categorical and a continuous variable
df1 = pd.DataFrame({'x':['P', 'Q', 'R', 'S', 'T', 'U'],'y':[10, 25, 35, 33, 41, 59]})
sns.barplot(x='x', y='y', data=df1)
plt.show()
```

```python
# distribution plot is a combination of a histogram with the default bin size 
# and a Kernel Density Estimation (KDE) plot
sns.distplot(df.satisfaction_level)
plt.show()
```

```python
# box plot, aka box-whisker plot, is one of the best plots to understand the distribution 
# of each variable with its quartiles
# it also shows the minimum and maximum and outliers in the data
sns.boxplot(data=df[['satisfaction_level', 'last_evaluation']])
plt.show()
```

```python
# the kde() function plots the probability density estimation of a given continuous variable
sns.kdeplot(df.satisfaction_level)
plt.show()
```

```python
# violin plots are a combined form of box plots and KDE, which offer 
# easy-to-understand analysis of the distribution
sns.violinplot(data=df[['satisfaction_level','last_evaluation']])
plt.show()
# here, we can conclude that the range of satisfaction_level is higher 
# than last_evaluation (performance) and both the variables have two peaks in the distribution
```

```python
# countplot() shows the frequency of each categorical variable 
# It is also known as a histogram for categorical variables
sns.countplot(x='salary', data=df)
plt.show()
```

```python
# we can also use hue as the second variable
sns.countplot(x='salary', data=df, hue='left')
plt.show()
```

```python
# the joint plot shows the bivariate relationship and distribution of individual variables in a single graph
sns.jointplot(x='satisfaction_level', y='last_evaluation', data=df)
plt.show()
```

```python
# by setting the kind parameter as "kde", we can draw the KDE plot
sns.jointplot(x='satisfaction_level', y='last_evaluation', data=df, kind='kde')
plt.show()
```

```python
# heatmap offers two-dimensional grid representation
# the individual cell of the grid contains a value of the matrix
iris_data = sns.load_dataset('iris')
corr_matrix = iris_data.corr()
sns.heatmap(data=corr_matrix, annot=True)
plt.show()
```

```python
# set a new color map using the cmap parameter
sns.heatmap(cor_matrix, annot=True, cmap="YlGnBu")
plt.show()
```

```python
# seaborn offers quick exploratory data analysis with relationships and individual distribution using a pair plot 
# a pair plot offers a single distribution using a histogram and joint distribution using a scatter plot
sns.pairplot(iris_data)
plt.show()
# the diagonal shows the distribution of the columns 
# the non-diagonal elements of the grid show the scatter plot to understand the relationship among all the variables.
```

### Interactive visualization with Bokeh

Bokeh is an interactive, high-quality, versatile, focused, and more powerful visualization library for large-volume and streaming data. It offers interactive, rich charts, plots, layouts, and dashboards for modern web browsers. Its output can be mapped to a notebook, HTML, or server.

Both the Matplotlib and Bokeh libraries have different intentions. Matplotlib focuses on static, simple, and fast visualization while Bokeh focuses on highly interactive, dynamic, web-based, and quality visualization. Matplotlib is generally used for publication images while Bokeh is for a web audience.

```python
# the output_notebook() function defines that the plot will render on the Jupyter Notebook
# the figure object is used as one of the core objects to draw charts and graphs
# the figure object focuses on the plot title, size, label, grids, and style
# the figure object also deals with plot style, title, axes labels, axes, grids, 
# and various methods for adding data
from bokeh.plotting import figure
from bokeh.plotting import output_notebook
from bokeh.plotting import show
```

```python
x = [1,3,5,7,9,11]
y = [10,25,35,33,41,59]

# define that the plot will render on the Jupyter Notebook
output_notebook()
# instantiate a figure
fig = figure(plot_width=500, plot_height=350)
# create scatter circle marker plot by rendering the circles
fig.line(x, y, line_width=3, color='orange')
fig.circle(x, y, size=10, color='green', alpha=1)
show(fig)
```

```python
fig.line(x, y, line_width=3, color='orange')
show(fig)
```

Bokeh offers layouts for organizing plots and widgets. Layouts organize more than one plot in a single panel for interactive visualizations. They also allow setting the sizing modes for resizing the plots and widgets based on panel size. The layout can be of the following types:

- *Row layout*: This organizes all the plots in a row or in a horizontal fashion.
- *Column layout*: This organizes all the plots in a column or in a vertical fashion.
- *Nested layout*: This is a combination of row and column layouts. 
- *Grid layout*: This offers you a grid of matrices for arranging the plots in.

```python
from bokeh.layouts import row, column
from bokeh.sampledata.iris import flowers as df
```

```python
fig1 = figure(plot_width=300, plot_height=300)
fig2 = figure(plot_width=300, plot_height=300)
fig3 = figure(plot_width=300, plot_height=300)
fig1.circle(df['petal_length'], df['sepal_length'], size=8, color='red', alpha=0.5)
fig2.circle(df['petal_length'], df['sepal_width'], size=8, color='green', alpha=0.5)
fig3.circle(df['petal_length'], df['petal_width'], size=8, color='blue', alpha=0.5)

# create row layout
row_layout = row(fig1, fig2, fig3)
show(row_layout)
```

```python
# create column layout
col_layout = column(fig1, fig2, fig3)
show(col_layout)
```

```python
# create nested layout
nested_layout = row(fig1, column(fig2, fig3))
show(nested_layout)
```

```python
# multiple plots
from bokeh.layouts import gridplot

grid_layout = gridplot([[fig1, fig2], [fig3, None]])
show(grid_layout)
```

```python
# hide click policy
from bokeh.models import CategoricalColorMapper

fig = figure(plot_width=500, plot_height=350, 
             title='Petal length Vs. Petal Width', 
             x_axis_label='petal_length', y_axis_label='petal_width')

# create scatter marker plot by render the circles 
for specie, color in zip(['setosa', 'virginica', 'versicolor'], ['blue', 'green', 'red']):
    data = df[df.species==specie]
    fig.circle('petal_length', 'petal_width', 
               size=8, color=color, alpha = 0.7, 
               legend_label=specie, source=data)

# set the legend location and click policy
fig.legend.location = 'top_left'
fig.legend.click_policy='hide'
show(fig)
```

```python
# mute click policy

fig = figure(plot_width=500, plot_height=350, 
             title='Petal length Vs. Petal Width', 
             x_axis_label='petal_length', y_axis_label='petal_width')

for specie, color in zip(['setosa', 'virginica', 'versicolor'], ['blue', 'green', 'red']):
    data = df[df.species==specie]
    fig.circle('petal_length', 'petal_width', 
               size=8, color=color, alpha = 0.7, 
               legend_label=specie, source=data,
               muted_color=color, muted_alpha=0.2)

# Set the legend location and click policy
fig.legend.location = 'top_left'
fig.legend.click_policy='mute'
show(fig)
```

```python
# annotations
# create color mapper for categorical column 
color_mapper = CategoricalColorMapper(factors=['setosa', 'virginica', 'versicolor'], 
                                      palette=['cyan', 'magenta', 'yellow'])

color_dict={'field': 'species','transform': color_mapper }
p = figure(plot_width = 500, plot_height = 350, title="Petal length Vs. Petal Width", 
           x_axis_label='petal_length', y_axis_label='petal_width')

p.circle('petal_length', 'petal_width', size=8, color=color_dict, alpha = 0.5, 
         legend_group='species', source=df)

p.legend.location = 'top_left'

show(p)
```

```python
# hover tool
from bokeh.models import HoverTool

mapper = CategoricalColorMapper(factors=['setosa', 'virginica', 'versicolor'], 
                                palette=['cyan', 'magenta', 'yellow'])

color_dict={'field': 'species','transform': mapper}

# create hovertool and specify the hovering information
hover = HoverTool(tooltips=[('Species type','@species'),
                            ('IRIS Petal Length','@petal_length'),
                            ('IRIS Petal Width', '@petal_width')])

p = figure(plot_width = 500, plot_height = 350, title="Petal length Vs. Petal Width", 
           x_axis_label='petal_length', y_axis_label='petal_width',
           tools=[hover, 'pan', 'wheel_zoom'])
p.circle('petal_length', 'petal_width', size=8, 
         color=color_dict, alpha = 0.5,
         legend_group='species',source=df)
p.legend.location = 'top_left'
show(p)
```
