# Reading HTML tables 

from Pandas 1.x Cookbook - Second Edition https://subscription.packtpub.com/book/programming/9781839213106/3/ch03lvl1sec32/reading-json

In this recipe, we will scrape tables from the Wikipedia entry for The Beatles Discography.

https://en.wikipedia.org/wiki/The_Beatles_discography

```python
import pandas as pd
import numpy as np
```

```python
url = 'https://en.wikipedia.org/wiki/The_Beatles_discography'
dfs = pd.read_html(url)
len(dfs)
```

```python
dfs[2]
```

```python
dfs1 = pd.read_html(url, match='List of studio albums', na_values='-')
```

```python
len(dfs1)
```

```python
dfs1[0].columns
```

```python
dfs1[0]
```

```python
df = dfs1[0]
df.columns = ['Title', 'Release', 'UK', 'AUS', 'CAN', 'FRA', 'GER', 'NOR', 'US', 'Certifications']
df
```

```python
res = (
     df.pipe(
         lambda df_: df_[
             ~df_.Title.str.startswith("Released")
         ]
     )
     .assign(
         release_date=lambda df_: pd.to_datetime(
             df_.Release.str.extract(
                 r"Released: (.*) Label"
             )[0].str.replace(r"\[E\]", "")
         ),
         label=lambda df_: df_.Release.str.extract(
             r"Label: (.*)"
         ),
     )
     .loc[
         :,
         [
             "Title",
             "UK",
             "AUS",
             "CAN",
             "FRA",
             "GER",
             "NOR",
             "US",
             "release_date",
             "label",
         ],
     ]
 )
res
```

**How it works...**

The read_html function looks through the HTML for table tags and parses the contents into DataFrames. This can ease the scraping of websites. Unfortunately, as the example shows, sometimes data in HTML tables may be hard to parse. Rowspans and multiline headers may confuse pandas. You will want to make sure that you perform a sanity check on the result.

You can also use the attrs parameter to select a table from the page.

```python
url = 'https://github.com/mattharrison/datasets/blob/master/data/anscombes.csv'
dfs = pd.read_html(url, attrs={"class": "csv-data"})
dfs[0].head()
```

## Selecting the smallest of the largest

In this recipe, we find the five lowest budget movies from the top 100 scoring movies by taking advantage of the convenience methods: .nlargest and .nsmallest.

```python
movies = pd.read_csv(data_path+'movie.csv')
movies2 = movies[['movie_title', 'imdb_score', 'budget']]
movies2.head()
```

```python
# use the .nlargest method to select the top 100 movies by imdb_score
movies2.nlargest(100, 'imdb_score')
```

```python
# chain the .nsmallest method to return the five lowest budget films among those with a top 100 score
movies2.nlargest(100, 'imdb_score').nsmallest(5, 'budget')
```

## Selecting the largest of each group by sorting

One of the most basic and common operations to perform during data analysis is to select rows containing the largest value of some column within a group. For instance, this would be like finding the highest-rated film of each year or the highest-grossing film by content rating. 

```python
movies3 = movies[['movie_title', 'title_year', 'imdb_score']]
movies3.head()
```

```python
# use the .sort_values method to sort the DataFrame by title_year
# the default behavior sorts from the smallest to the largest
# use the ascending=True parameter to invert this behavior
movies3.sort_values('title_year', ascending=True)
```

```python
# to sort multiple columns at once, use a list
movies3.sort_values(['title_year', 'imdb_score'], ascending=False)
```

```python
# use the .drop_duplicates method to keep only the first row of every year
movies[['movie_title', 'title_year', 'imdb_score']] \
.sort_values(['title_year', 'imdb_score'], ascending=False) \
.drop_duplicates(subset='title_year')
```

It is possible to sort one column in ascending order while simultaneously sorting another column in descending order. To accomplish this, pass in a list of Booleans to the ascending parameter that corresponds to how you would like each column sorted. The following sorts title_year and content_rating in descending order and budget in ascending order. It then finds the lowest budget film for each year and content rating group:

```python
movies[['movie_title', 'title_year', 'content_rating', 'budget']] \
.sort_values(['title_year', 'content_rating', 'budget'], ascending=[False, False, True]) \
.drop_duplicates(subset=['title_year', 'content_rating'])
```

```python
def has_lucky_number(nums):
    """Return whether the given list of numbers is lucky. A lucky list contains
    at least one number divisible by 7.
    """
    for num in nums:
        if num % 7 == 0:
            return True
    return False
```

```python
has_lucky_number([5, 14, 3, 3])
```

```python
nums = [5, 14, 3, 3]
for num in nums:
        print(num)
        print(num % 7)
        if num % 7 == 0:
            print('lucky')
```
