# FIFA Football World Cup

Football/Soccer World Cup history from 1930 to 2022, all matches from [Kaggle](https://www.kaggle.com/datasets/piterfm/fifa-football-world-cup?select=matches_1930_2018.csv)

Possible questions to data:
1. Which countries hosted World Cup more than once?
2. Which countries took part in every World Cups?
3. How many times every countrie took part in the World Cup? Get the list of the years.
4. Who made the most of the goals in the World Cup history?

```python
headers = !head -n 1 data/world_cup/matches_1930_2022.csv
list(headers[0].split(','))
```

```python
df_wc = pd.read_csv('data/world_cup/matches_1930_2022.csv', parse_dates=True,
                usecols=['home_team', 'away_team', 'home_score', 
                         'home_penalty','away_penalty',
                         'away_score', 'home_goal', 'away_goal', 
                         'Venue', 'Date', 'Round', 'Host', 'Year'])
df_wc.columns = df_wc.columns.str.lower().str.replace(' ', '_')
df_wc.rename(columns={'round': 'stage'}, inplace=True)
df_wc.head(3)
```

```python
df_wc.home_team.unique()
```

```python
countries = {'West Germany': 'Germany',
             'Yugoslavia': 'Serbia',
             'FR Yugoslavia': 'Serbia',
             'Czechoslovakia': 'Czech Republic',
             'Soviet Union': 'Russia',
             'Germany DR': 'Germany'}
```

```python
df_wc.replace(countries, inplace=True)
```

```python
df_wc.home_team.unique()
```

```python
loc = df_wc.columns.get_loc('away_penalty') + 1
loc
```

```python
# insert a column 'winner' after the column 'away_penalty'

df_wc.insert(
    loc=loc,
    column='winner',
    value=np.where(df_wc.home_score > df_wc.away_score, df_wc.home_team, df_wc.away_team))
```

```python
idx = df_wc.query('stage == "Final" and home_score == away_score').index.to_list()
```

```python
df_wc.loc[idx] = df_wc.loc[idx]\
        .assign(
            winner = lambda x: np.where(x.home_penalty > x.away_penalty, x.home_team, x.away_team))
```
