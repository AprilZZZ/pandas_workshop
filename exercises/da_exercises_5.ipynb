```python
df = pd.DataFrame({
    'ord_no':[70001,np.nan,70002,70004,np.nan,70005,np.nan,70010,70003,70012,np.nan,70013],
    'purch_amt':[150.5,270.65,65.26,110.5,948.5,2400.6,5760,1983.43,2480.4,250.45, 75.29,3045.6],
    'ord_date': ['2012-10-05',
                 '2012-09-10',
                 np.nan,
                 '2012-08-17',
                 '2012-09-10',
                 '2012-07-27',
                 '2012-09-10',
                 '2012-10-10',
                 '2012-10-10',
                 '2012-06-27',
                 '2012-08-17',
                 '2012-04-25'],
    'customer_id':[3002,3001,3001,3003,3002,3001,3001,3004,3003,3002,3001,3001],
    'salesman_id':[5002,5003,5001,np.nan,5002,5001,5001,np.nan,5003,5002,5003,np.nan]})
df
```

```python
# 1 Write a Pandas program to detect missing values of a given DataFrame. Display True or False

df.isnull()
```

```python
# 2 Write a Pandas program to identify the column(s) of a given DataFrame which have at least one missing value

df.isna().any()
```

```python
# 3 Write a Pandas program to count the number of missing values in each column of a given DataFrame

df.isnull().sum()
```

```python
# 4 Write a Pandas program to find and replace the missing values in a given DataFrame 
# which do not have any valuable information
# Missing values: ?, --
# Replace those values with NaN

df2 = pd.DataFrame({
    'ord_no':[70001,np.nan,70002,70004,np.nan,70005,"--",70010,70003,70012,np.nan,70013],
    'purch_amt':[150.5,270.65,65.26,110.5,948.5,2400.6,5760,"?",12.43,2480.4,250.45, 3045.6],
    'ord_date': ['?',
                 '2012-09-10',
                 np.nan,
                 '2012-08-17',
                 '2012-09-10',
                 '2012-07-27',
                 '2012-09-10',
                 '2012-10-10',
                 '2012-10-10',
                 '2012-06-27',
                 '2012-08-17',
                 '2012-04-25'],
    'customer_id':[3002,3001,3001,3003,3002,3001,3001,3004,"--",3002,3001,3001],
    'salesman_id':[5002,5003,"?",5001,np.nan,5002,5001,"?",5003,5002,5003,"--"]})
df2
```

```python
df2.replace(['--', '?'], np.nan)
```

```python
# 5 Write a Pandas program to drop the rows where at least one element is missing in a given DataFrame
df.dropna()
```

```python
# 6 Write a Pandas program to drop the columns where at least one element is missing in a given DataFrame.

df.dropna(axis=1)
```

```python
# 7 Write a Pandas program to drop the rows where all elements are missing in a given DataFrame

df.loc[0, :] = np.nan
df.dropna(how='all')
```

```python
# 8 Write a Pandas program to keep the rows with at least 2 NaN values in a given DataFrame

df.dropna(thresh=2)
```

```python
# 9 Write a Pandas program to drop those rows from a given DataFrame in which specific columns 
# have missing values

df.dropna(subset=['ord_no', 'purch_amt', 'customer_id'])
```

```python
# 10 Write a Pandas program to keep the valid entries of a given DataFrame

df2 = df.copy()
df2.dropna(inplace=True)
df2
```

```python
# 11 Write a Pandas program to calculate the total number of missing values in a DataFrame

df.isnull().sum().sum()
```

```python
# 12 Write a Pandas program to replace NaNs with a single constant value in specified columns in a DataFrame

df.salesman_id.fillna(5001, inplace=True)
df
```

```python
# 13 Write a Pandas program to replace NaNs with the value from the previous row 
# or the next row in a given DataFrame

df.purch_amt.bfill(inplace=True)
df
```

```python
df.ord_no.ffill(inplace=True)
df
```

```python
# 14 Write a Pandas program to replace NaNs with median or mean of the specified columns in a given DataFrame

df.loc[2:7, 'purch_amt'] = np.nan
df.purch_amt.fillna(df.purch_amt.mean(), inplace=True)
df
```

```python
# 15 Write a Pandas program to interpolate the missing values using the Linear Interpolation method

df.loc[2:7, 'purch_amt'] = np.nan
df.purch_amt.interpolate(inplace=True)
df
```

```python
# 16 Write a Pandas program to count the number of missing values of a specified column

df.ord_date.isnull().sum()
```

```python
df.ord_date.value_counts(dropna=False).loc[np.nan]
```

```python
# 17 Write a Pandas program to count the missing values 

df.loc[2:7, 'purch_amt'] = np.nan
df.isnull().sum().sum()
```

```python
# 18 Write a Pandas program to find the Indexes of missing values in a given DataFrame

df.purch_amt.isnull().to_numpy().nonzero()
```

```python
# 19 Write a Pandas program to replace the missing values with the most frequent values present in each column

df.loc[3:7, 'salesman_id'] = np.nan
most_active_salesman = df.salesman_id.value_counts().idxmax()
df.salesman_id.fillna(most_active_salesman, inplace=True)
df
```

```python
# 20 Write a Pandas program to create a hitmap for more information about the distribution of missing values

df = pd.DataFrame({
    'ord_no':[70001,np.nan,70002,70004,np.nan,70005,np.nan,70010,70003,70012,np.nan,70013],
    'purch_amt':[150.5,np.nan,65.26,110.5,948.5,np.nan,5760,1983.43,np.nan,250.45, 75.29,3045.6],
    'sale_amt':[10.5,20.65,np.nan,11.5,98.5,np.nan,57,19.43,np.nan,25.45, 75.29,35.6],
'ord_date': ['2012-10-05',
             '2012-09-10',
             np.nan,
             '2012-08-17',
             '2012-09-10',
             '2012-07-27',
             '2012-09-10',
             '2012-10-10',
             '2012-10-10',
             '2012-06-27',
             '2012-08-17',
             '2012-04-25'],
    'customer_id':[3002,3001,3001,3003,3002,3001,3001,3004,3003,3002,3001,3001],
    'salesman_id':[5002,5003,5001,np.nan,5002,5001,5001,np.nan,5003,5002,5003,np.nan]})
df
```
