```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Pandas Excel [25 exercises with solution]

```python
# 1 Write a Pandas program to import excel data (coalpublic2013.xlsx ) into a Pandas dataframe

df_ex = pd.read_excel('files/coalpublic2013.xlsx')
df_ex.head(5)
```

```python
# 2 Write a Pandas program to get the data types of the given excel data (coalpublic2013.xlsx ) fields

df_ex.dtypes
```

```python
# 3 Write a Pandas program to read specific columns from a given excel file

df_ex_1 = pd.read_excel('files/coalpublic2013.xlsx', usecols=['MSHA ID', 'Mine_Name', 'Labor_Hours'])
df_ex_1.head(3)
```

```python
# 4 Write a Pandas program to find the sum, mean, max, min value of 'Production (short tons)' 
# column of coalpublic2013.xlsx file

df_ex.Production.agg(['sum', 'mean', 'max', 'min'])
```

```python
# 5 Write a Pandas program to insert a column in the sixth position of the said excel sheet 
# and fill it with NaN values

df_ex.insert(5, 'new_col', np.nan)
df_ex.to_excel('files/new_file.xlsx', sheet_name='new_sheet', na_rep='NaN')
```

```python
# 6 Write a Pandas program to import some excel data (coalpublic2013.xlsx ) 
# skipping first twenty rows into a Pandas dataframe

df_ex_2 = pd.read_excel('files/coalpublic2013.xlsx', skiprows=20)
df_ex.shape, df_ex_2.shape
```

```python
# 9 Write a Pandas program to create a subtotal of "Labor Hours" against MSHA ID from the given excel data

df_ex[['MSHA ID', 'Labor_Hours']].groupby('MSHA ID').sum().head(10)
```

```python
# 12  Write a Pandas program to find details where "Mine Name" starts with "P" 

df_ex[df_ex.Mine_Name.str.startswith('P')]
```

```python
# 14 Write a Pandas program to import excel data and find a list of specified customers by name

df_ex.query('Mine_Name in ("Shoal Creek Mine", "Piney Woods Preparation Plant")')
```

```python
# 15 Write a Pandas program to import excel data (employee.xlsx) into a Pandas dataframe 
# and find a list of employees where hire_date> 01-01-07

df_emp = pd.read_excel('files/employee.xlsx')
df_emp.query('hire_date > "01-01-07"')
```

```python
# 16 Write a Pandas program to import excel data and to sort the records by the hire_date column

df_emp.sort_values('hire_date').head(3)
```

```python
# 18 Write a Pandas program to import given excel data and find a list of employees of a specified year

df_emp.set_index('hire_date').loc['2005']
```

```python
# 21 Write a Pandas program to import sheet2 data 

df_emp_2 = pd.read_excel('files/employee.xlsx', sheet_name='Sheet2')
df_emp_2.head(5)
```

```python
# 22 Write a Pandas program to import three datasheets from a given excel data and combine in to a single dataframe

df_emp_full = pd.concat(pd.read_excel('files/employee.xlsx', sheet_name=None), ignore_index= True)
df_emp_full                                                     
```
