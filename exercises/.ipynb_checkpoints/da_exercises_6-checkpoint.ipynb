# Pandas Pivot Table [13 exercises with solution]

```python
df = pd.read_excel('files/SaleData.xlsx')
df.head(5)
```

```python
# 1 Write a Pandas program to create a Pivot table with multiple indexes from a given excel sheet (Salesdata.xlsx)

pd.pivot_table(
    df,
    index=['Region', 'Manager'],
    columns=['Item'],
    values=['Units', 'Sale_amt'],
    aggfunc=np.sum)
```

```python
pd.pivot_table(
    df,
    index=['Region', 'SalesMan'])
```

```python
# 2 Write a Pandas program to create a Pivot table and find the total sale amount region wise, manager wise

pd.pivot_table(
    df,
    index=['Region', 'Manager'],
    values=['Sale_amt'],
    aggfunc=np.sum)
```

```python
# 3 Write a Pandas program to create a Pivot table and find the total sale amount region wise, 
# manager wise, sales man wise

pd.pivot_table(
    df,
    index=['Region', 'Manager', 'SalesMan'],
    values=['Sale_amt'],
    aggfunc=np.sum)
```

```python
# 4 Write a Pandas program to create a Pivot table and find the item wise unit sold

pd.pivot_table(
    df,
    index=['Item'],
    values=['Units'],
    aggfunc=np.sum)
```

```python
# 5 Write a Pandas program to create a Pivot table and find the region wise total sale

pd.pivot_table(
    df,
    index=['Region'],
    values=['Sale_amt'],
    aggfunc=np.sum)
```

```python
# 6 Write a Pandas program to create a Pivot table and find the region wise, item wise unit sold

pd.pivot_table(
    df,
    index=['Region', 'Item'],
    values=['Units'],
    aggfunc=np.sum)
```

```python
# 7 Write a Pandas program to create a Pivot table and count the manager wise sale and mean value of sale amount

pd.pivot_table(
    df,
    index=['Manager'],
    values=['Sale_amt'],
    aggfunc=[np.mean, 'count'])
```

```python
# 8 Write a Pandas program to create a Pivot table and find manager wise, salesman wise total sale 
# and also display the sum of all sale amount at the bottom

pd.pivot_table(
    df,
    index=['Manager', 'SalesMan'],
    values=['Units', 'Sale_amt'],
    aggfunc=np.sum,
    fill_value=0,
    margins=True)
```

```python
# 9 Write a Pandas program to create a Pivot table and find the total sale amount region wise, 
# manager wise, sales man wise where Manager = "Douglas"

pd.pivot_table(
    df,
    index=['Region', 'Manager', 'SalesMan'],
    values=['Sale_amt'],
    aggfunc=np.sum).query('Manager == "Douglas"')
```

```python
# 10 Write a Pandas program to create a Pivot table and find the region wise Television and Home Theater sold

pd.pivot_table(
    df.query('Item in ["Television", "Home Theater"]'),
    index=['Region'],
    columns=['Item'],
    values=['Units'],
    aggfunc=np.sum)
```

```python
# 11 Write a Pandas program to create a Pivot table and find the maximum sale value of the items

pd.pivot_table(
    df,
    index=['Item'],
    values=['Sale_amt'],
    aggfunc=np.max)
```

```python
# 12 Write a Pandas program to create a Pivot table and find the minimum sale value of the items

pd.pivot_table(
    df,
    index=['Item'],
    values=['Sale_amt'],
    aggfunc=np.min)
```

```python
# 13 Write a Pandas program to create a Pivot table and find the maximum and minimum sale value of the items

pd.pivot_table(
    df,
    index=['Item'],
    values=['Sale_amt'],
    aggfunc=[np.min, np.max])
```

# Pandas Pivot on Titanic Passengers csv [19 exercises with solution]

```python
df_t = pd.read_csv('files/titanic.csv')
df_t.head(5)
```

```python
# 1 Write a Pandas program to print a concise summary of the dataset

df_t.info()
```

```python
# 2 Write a Pandas program to extract the column labels, shape and data types of the dataset

print(f'columns: {df_t.columns.to_list()}')
print(f'shape: {df_t.shape}')
print(f'data types: {df_t.dtypes}')
```

```python
# 3 Write a Pandas program to create a Pivot table with multiple indexes from the data set 

pd.pivot_table(
    df_t,
    index=['class', 'sex'],
    values=['survived'],
    aggfunc='count')
```

```python
# 4 Write a Pandas program to create a Pivot table and find survival rate by gender on various classes

df_t.pivot_table(
    values='survived',
    index='sex',
    columns='class')
```

```python
# 5 Write a Pandas program to create a Pivot table and find survival rate by gender

df_t.pivot_table(
    values='survived',
    index='sex')
```

```python
# 6 Write a Pandas program to create a Pivot table and find survival rate by gender, age wise of various classes

df_t.pivot_table(
    values='survived',
    index=['sex', 'age'],
    columns=['class'])
```

```python
# 7 Write a Pandas program to partition each of the passengers into four categories based on their age
# Note: Age categories (0, 10), (10, 30), (30, 60), (60, 80)

bins = pd.IntervalIndex.from_tuples([(0, 10), (10, 30), (30, 60), (60, 80)])
age_categories = pd.cut(df_t.age, bins=bins)
age_categories.value_counts()
```

```python
# 8 Write a Pandas program to create a Pivot table and count survival by gender, categories 
# wise age of various classes
# Note: Age categories (0, 10), (10, 30), (30, 60), (60, 80)

df_t['age_category'] = pd.cut(df_t.age, bins=bins)
df_t.pivot_table(
    values='survived',
    index=['sex', 'age_category'],
    columns=['class'],
    aggfunc='count'
)
```

```python
# 9 Write a Pandas program to create a Pivot table and find survival rate by gender, 
# age of the different categories of various classes

df_t['age_cat'] = pd.cut(df_t.age, [0, 20, 50])
df_t.pivot_table(
    values='survived',
    index=['sex', 'age_cat'],
    columns=['class']
)
```

```python
# 10 Write a Pandas program to create a Pivot table and find survival rate by gender, 
# age of the different categories of various classes. 
# Add the fare as a dimension of columns and partition fare column into 2 categories based 
# on the values present in fare columns.

df_t['fare_cat'] = pd.cut(df_t.fare, 2)
df_t.pivot_table(
    values='survived',
    index=['sex', 'age_category'],
    columns=['class', 'fare_cat'])
```

```python
# 11 Write a Pandas program to create a Pivot table and calculate number of women and men
# were in a particular cabin class

df_t.pivot_table(
    values='survived',
    index=['sex'],
    columns=['class'],
    aggfunc='count',
    margins=True)
```

```python
# 12 Write a Pandas program to create a Pivot table and find survival of both gender and class affected

df_t.groupby(['sex', 'class']).survived.mean().unstack()
```

```python
# 14 Write a Pandas program to create a Pivot table and calculate how many women and men 
# were in a particular cabin class

df_t.pivot_table(
    values=['survived'],
    index=['sex'],
    columns=['class'])
```

```python
# 15 Write a Pandas program to create a Pivot table and find number of survivors and average rate 
# grouped by gender and class

df_t.pivot_table(
    values=['survived'],
    index=['sex', 'class'],
    aggfunc=[np.mean, 'count']).unstack()
```

```python
# 16 Write a Pandas program to create a Pivot table and find number of adult male, adult female and children

df_t.pivot_table(
    values='sex',
    index='who',
    aggfunc='count')
```

```python
# 17 Write a Pandas program to create a Pivot table and check missing values of children

df_t.loc[df_t.who == 'child'].isnull().sum()
```

```python
# 18 Write a Pandas program to create a Pivot table and separate the gender according to whether 
# they traveled alone or not to get the probability of survival

df_t.pivot_table(
    values='survived',
    index=['sex', 'alone'],
    columns=['class'])
```

```python
# 19 Write a Pandas program to create a Pivot table and find the probability of survival by class, 
# gender, solo boarding and port of embarkation

df_t.pivot_table(
    values=['survived'],
    index=['class', 'sex', 'alone'],
    columns=['embark_town'],
    aggfunc=['count', 'mean'])
```
