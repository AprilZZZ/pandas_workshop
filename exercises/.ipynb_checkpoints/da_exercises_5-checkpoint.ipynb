# Pandas Handling Missing Values [ 20 exercises with solution]

```python
df = pd.DataFrame({
    'ord_no':[70001,np.nan,70002,70004,np.nan,70005,np.nan,70010,70003,70012,np.nan,70013],
    'purch_amt':[150.5,270.65,65.26,110.5,948.5,2400.6,5760,1983.43,2480.4,250.45, 75.29,3045.6],
    'ord_date': ['2012-10-05',
                 '2012-09-10',
                 np.nan,
                 '2012-08-17',
                 '2012-09-10',
                 '2012-07-27',
                 '2012-09-10',
                 '2012-10-10',
                 '2012-10-10',
                 '2012-06-27',
                 '2012-08-17',
                 '2012-04-25'],
    'customer_id':[3002,3001,3001,3003,3002,3001,3001,3004,3003,3002,3001,3001],
    'salesman_id':[5002,5003,5001,np.nan,5002,5001,5001,np.nan,5003,5002,5003,np.nan]})
df
```

```python
# 1 Write a Pandas program to detect missing values of a given DataFrame. Display True or False

df.isnull()
```

```python
# 2 Write a Pandas program to identify the column(s) of a given DataFrame which have at least one missing value

df.isna().any()
```

```python
# 3 Write a Pandas program to count the number of missing values in each column of a given DataFrame

df.isnull().sum()
```

```python
# 4 Write a Pandas program to find and replace the missing values in a given DataFrame 
# which do not have any valuable information
# Missing values: ?, --
# Replace those values with NaN

df2 = pd.DataFrame({
    'ord_no':[70001,np.nan,70002,70004,np.nan,70005,"--",70010,70003,70012,np.nan,70013],
    'purch_amt':[150.5,270.65,65.26,110.5,948.5,2400.6,5760,"?",12.43,2480.4,250.45, 3045.6],
    'ord_date': ['?',
                 '2012-09-10',
                 np.nan,
                 '2012-08-17',
                 '2012-09-10',
                 '2012-07-27',
                 '2012-09-10',
                 '2012-10-10',
                 '2012-10-10',
                 '2012-06-27',
                 '2012-08-17',
                 '2012-04-25'],
    'customer_id':[3002,3001,3001,3003,3002,3001,3001,3004,"--",3002,3001,3001],
    'salesman_id':[5002,5003,"?",5001,np.nan,5002,5001,"?",5003,5002,5003,"--"]})
df2
```

```python
# 5 Write a Pandas program to drop the rows where at least one element is missing in a given DataFrame
df.dropna()
```

```python
# 6 Write a Pandas program to drop the columns where at least one element is missing in a given DataFrame.

df.dropna(axis=1)
```

```python
# 7 Write a Pandas program to drop the rows where all elements are missing in a given DataFrame

df.loc[0, :] = np.nan
df.dropna(how='all')
```

```python
# 8 Write a Pandas program to keep the rows with at least 2 NaN values in a given DataFrame

df.dropna(thresh=2)
```

```python
# 9 Write a Pandas program to drop those rows from a given DataFrame in which specific columns 
# have missing values

df.dropna(subset=['ord_no', 'purch_amt', 'customer_id'])
```

```python
# 10 Write a Pandas program to keep the valid entries of a given DataFrame

df2 = df.copy()
df2.dropna(inplace=True)
df2
```

```python
# 11 Write a Pandas program to calculate the total number of missing values in a DataFrame

df.isnull().sum().sum()
```

```python
# 12 Write a Pandas program to replace NaNs with a single constant value in specified columns in a DataFrame

df.salesman_id.fillna(5001, inplace=True)
df
```

```python
# 13 Write a Pandas program to replace NaNs with the value from the previous row 
# or the next row in a given DataFrame

df.purch_amt.bfill(inplace=True)
df
```

```python
df.ord_no.ffill(inplace=True)
df
```

```python
# 14 Write a Pandas program to replace NaNs with median or mean of the specified columns in a given DataFrame

df.loc[2:7, 'purch_amt'] = np.nan
df.purch_amt.fillna(df.purch_amt.mean(), inplace=True)
df
```

```python
# 15 Write a Pandas program to interpolate the missing values using the Linear Interpolation method

df.loc[2:7, 'purch_amt'] = np.nan
df.purch_amt.interpolate(inplace=True)
df
```

```python
# 16 Write a Pandas program to count the number of missing values of a specified column

df.ord_date.isnull().sum()
```

```python
df.ord_date.value_counts(dropna=False).loc[np.nan]
```

```python
# 17 Write a Pandas program to count the missing values 

df.loc[2:7, 'purch_amt'] = np.nan
df.isnull().sum().sum()
```

```python
# 18 Write a Pandas program to find the Indexes of missing values in a given DataFrame

df.purch_amt.isnull().to_numpy().nonzero()
```

```python
# 19 Write a Pandas program to replace the missing values with the most frequent values present in each column

df.loc[3:7, 'salesman_id'] = np.nan
most_active_salesman = df.salesman_id.value_counts().idxmax()
df.salesman_id.fillna(most_active_salesman, inplace=True)
df
```

# Pandas Styling [15 exercises with solution]

```python
# create a dataframe of ten rows, four columns with random values 

np.random.seed(24)
df_a = pd.DataFrame({'A': np.linspace(1, 10, 10)})
df_s = pd.concat([df_a, pd.DataFrame(np.random.randn(10, 4), columns=list('BCDE'))], axis=1)
df_s
```

```python
# 1 Write a Pandas program to highlight the negative numbers red and positive numbers black

df_s.style.applymap(lambda x: 'color: red' if x < 0 else 'color: black')
```

```python
# 2 convert some values to nan values, write a Pandas program which will highlight the nan values

df_s.iloc[0, 2] = np.nan
df_s.iloc[3, 3] = np.nan
df_s.iloc[4, 1] = np.nan
df_s.iloc[9, 4] = np.nan
df_s.style.highlight_null(null_color='red')
```

```python
# 3 write a Pandas program to highlight the maximum value in each column

df_s.style.highlight_max(subset=['B', 'C', 'D', 'E'], color='green')
```

```python
# 4 write a Pandas program to highlight the minimum value in each column

df_s.style.highlight_min(subset=['B', 'C', 'D', 'E'], color='yellow')
```

```python
# 5 write a Pandas program to highlight the maximum value in last two columns

df_s.style.highlight_max(subset=['D', 'E'], color='green')
```

```python
# 6 write a Pandas program to set dataframe background Color black and font color yellow

df_s.style.set_properties(**{'background-color': 'black', 'color': 'yellow'})
```

```python
# 7 write a Pandas program to highlight dataframe's specific columns

df_s.style.set_properties(subset=['A', 'B'], **{'background-color': 'grey'})
```

```python
# 8 write a Pandas program to highlight dataframe's specific columns with different colors

df_s.style.set_properties(subset=['B', 'C', 'E'], **{'background-color': 'grey'})\
        .set_properties(subset=['A', 'D'], **{'background-color': 'red'})
```

```python
# 9 write a Pandas program to display the dataframe in table style

# CSS properties for th elements in dataframe
th_props = [
    ('font-size', '14px'),
    ('text-align', 'center'),
    ('font-weight', 'bold'),
    ('color', '#6d6d6d'),
    ('background-color', '#f7ffff')
]

# CSS properties for td elements in dataframe
td_props = [('font-size', '10px')]

# table styles
styles = [
    dict(selector="th", props=th_props),
    dict(selector="td", props=td_props)
]

df_s.style.set_table_styles(styles)
```

```python
# 10 write a Pandas program to highlight the entire row in Yellow 
# where a specific column value is greater than 0.5

df_s.style.apply(
    lambda x: ['background-color: yellow']*5 
        if x.C > 0.5 
        else ['background-color: white']*5, 
    axis=1)
```

```python
# 11 write a Pandas program to display the dataframe in Heatmap style

df_s.style.background_gradient(cmap='viridis')
```

```python
# 12 write a Pandas program to make a gradient color mapping on a specified column

df_s.style.background_gradient(subset=['C'])
```

```python
# 13 write a Pandas program to make a gradient color on all the values of the said dataframe

df_s.style.background_gradient()
```

```python
# 14 Write a Pandas program to display the dataframe in table style and border around the table 
# and not around the rows

df_s.style.set_table_styles([
    {
        'selector': '', 
        'props': [('border','4px solid #7a7')]
    }
])
```

```python
# 15 write a Pandas program to display bar charts in dataframe on specified columns

df_s.style.bar(subset=['B', 'C'], color='#d65f5f')
```
