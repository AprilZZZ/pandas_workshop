This notebook contains the forth part of exercises from: https://www.w3resource.com/python-exercises/pandas/index.php It covers filtering, grouping and aggregating data.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('alco_ds.csv')
df.head()
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
df.head()
```

# Pandas Filter [ 27 exercises with solution ]

```python
# 1 Write a Pandas program to display the dimensions or shape of the World alcohol consumption dataset. 
# Also extract the column names from the dataset.

df.info()
```

```python
# 2 Write a Pandas program to select first 2 rows, 2 columns and specific two columns

df.iloc[:, :2].head(2)
```

```python
# 3 Write a Pandas program to select random number of rows, fraction of random rows

df.sample(5)
df.sample(frac=0.1)
```

```python
# 4 Write a Pandas program to find and drop the missing values from World alcohol consumption dataset.

df.isna().sum()
df1 = df.dropna()
df1.isna().sum()
```

```python
# 5 Write a Pandas program to remove the duplicates from 'WHO region' column of World alcohol consumption dataset.

df1[df1.who_region.duplicated()]
df2 = df1.drop_duplicates('who_region')
df2
```

```python
# 6  Write a Pandas program to find out the alcohol consumption of a given year

df1.groupby('year').sum()
df1[df1.year == 1984].display_value.sum()
```

```python
# 7 Write a Pandas program to find out the alcohol consumption details in the year '1987' or '1989'

df1.query('year == 1987 or year == 1989')
```

```python
# 8 Write a Pandas program to find out the alcohol consumption details by the 'Americas' in the year '1989'

df1.query('year == 1989 and who_region == "Americas"')
```

```python
# 9 Write a Pandas program to find out the alcohol consumption details in the year '1986' 
# where WHO region is 'Western Pacific' and country is 'VietNam'

df1.query('year == 1986 and who_region == "Western Pacific" and country == "Viet Nam"')
```

```python
# 10 Write a Pandas program to find out the alcohol consumption details in the year '1986' or '1989' 
# where WHO region is 'Americas'

df1.query('(year == 1986 or year == 1989) and who_region == "Americas"')
```

```python
# 11 Write a Pandas program to find out the alcohol consumption details in the year '1986' or '1989' 
# where WHO region is 'Americas' or 'Europe'

df1.query('year in (1986, 1989) and who_region in ("Americas", "Europe")')
```

```python
# 12 Write a Pandas program to find out the 'WHO region, 'Country', 'Beverage Types' 
# in the year '1986' or '1989' where WHO region is 'Americas' or 'Europe'

df1.query('year in (1986, 1989) and who_region in ("Americas", "Europe")')\
    [['who_region', 'country', 'beverage_types']]
```

```python
# 13 Write a Pandas program to find out the records where consumption of beverages 
# per person average >=5 and Beverage Types is Beer

df1.query('display_value >= 5 and beverage_types == "Beer"')
```

```python
# 14 Write a Pandas program to find out the records where consumption of beverages 
# per person average >=4 and Beverage Types is Beer, Wine, Spirits
df1.query('display_value >=4 and beverage_types in ("Beer", "Wine", "Spirits")')
```

```python
# 15 Write a Pandas program to filter the specified columns and records by range

df1[df1.year.between(1984, 1987) & df1.display_value.between(4, 9)]
```

```python
df1.loc[0:4, ['who_region', 'beverage_types']]
```

```python
# 16 Write a Pandas program to filter those records where WHO region contains "Ea" substring

df1[df1.who_region.str.contains('Ea')]
```

```python
# 17 Write a Pandas program to filter those records where WHO region matches with multiple values 
# (Africa, Eastern Mediterranean, Europe) 

df1.query('who_region in ("Africa", "Eastern Mediterranean", "Europe")')
```

```python
# 18 Write a Pandas program to filter those records which not appears in a given list 

df1.query('who_region not in ("Africa", "Eastern Mediterranean", "Europe")')
```

```python
# 19 Write a Pandas program to filter all records where the average consumption of beverages 
# per person from .5 to 2.50

df1[df1.display_value.between(.5, 2.50)]
```

```python
# 20 Write a Pandas program to find average consumption of wine per person greater than 2

df1.query('beverage_types == "Wine" and display_value > 2')
```

```python
# 21 Write a Pandas program to filter rows based on row numbers ended with 0, like 0, 10, 20, 30 

df1.reset_index().iloc[::10]
```

```python
df1.filter(regex='0$', axis=0)
```

```python
# 22 Write a Pandas program to select consecutive columns and also select rows 
# with Index label 0 to 9 with some columns

df1.loc[0:9, 'country':'display_value']
```

```python
# 24 Write a Pandas program to find which years have all non-zero values and which years have any non-zero values

df1.loc[:, df1.all()]
df1.loc[:, df1.any()]
```

```python
# 25 Write a Pandas program to filter all columns where all entries present, 
# check which rows and columns has a NaN and finally drop rows with any NaNs

df.loc[:, df.notnull().all()]
df.loc[:, df.isnull().any()]
```

```python
# 26 Write a Pandas program to filter all records starting from the 'Year' column, access every other column

df1.loc[:, 'year'::2]
```

```python
# 27 Write a Pandas program to filter all records starting from the 2nd row, access every 5th row

df1.loc[1::5]
```
